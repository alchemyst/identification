```python
import pandas
import lvm
import fit
```

```python
filename = 'raw/SmallThin_Copper_ThinPadR1.LVM'
indata = lvm.lvm(filename)
response = fit.responsedata(indata.data.X_Value, indata.data.Voltage_0, indata.data.Voltage, name=f)
```

```python
Nterms = 100
L = 25.0e-3 # Length of sample (m)
c = 1.0 # Pulse height (W/m^2)
alpha = 1.11e-4 # Thermal diffusivity (m^2/s). Value for copper 1.11e-4 http://en.wikipedia.org/wiki/Thermal_diffusivity
alpha = 1.11e-4
beta = 7.0 # Pulse length (seconds)

t = linspace(0, 60, 100)
f = zeros(len(t))

for k in arange(1, Nterms+1):
    wk = (2*k - 1)*numpy.pi/(2.0*L)
    wk2 = wk**2
    Ak = 2*c**2/(L*wk2)
    k = 2*c**2/(L*wk)
    term = where(t<beta, 
                 k*(1 - numpy.exp(-alpha*wk2*t)),
                 k*(1 - numpy.exp(-alpha*wk2*beta))*numpy.exp(-alpha*wk2*(t-beta)))
    f += term
plot(t+38, f/max(f), response.t, response.y/max(response.y))
xlabel('Time')
ylabel('Flux')
xlim([30, 150])
savefig('Sample_solution.pdf')
```
