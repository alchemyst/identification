## Convective boundary condition

```python
import scipy.optimize
```

```python
L = 1.
alpha = 1.
k = 3.
h = 2.
n = 1.

def lambda_tan(lam):
    return lam*k/h*tan(lam*L) - 1

def lambda_sin(lam):
    return lam*sin(lam*L) - h/k*cos(lam*L)

def lambda_eq(lam):
    return sin(lam*L) - h/k*cos(lam*L)/lam

Nterms = 10
t = atleast_2d(linspace(0, 1)).T
x = linspace(0., L, 10)
flux = -ones((len(t), len(x)))
T = zeros((len(t), len(x))) + L - x
fluxkernel = zeros_like(t)

lambda_n = -pi/L/2.0
l = linspace(0, Nterms*pi/L, 1000)
plot(l, lambda_tan(l), label=r'$\lambda \frac{k}{h} \tan(\lambda L) - 1$')
plot(l, lambda_sin(l), label=r'$\lambda \sin(\lambda L) - \frac{h}{k}\cos(\lambda L)$')
plot(l, lambda_eq(l), label=r'$\sin(\lambda L) - \frac{h}{k \lambda}\cos(\lambda L)$')
axhline(0)
ylim([-5, 5])

lambda_n = -pi/L/2.
for i in range(1, Nterms):
    lambda_n = scipy.optimize.bisect(lambda_eq, lambda_n+pi/L/2., lambda_n+pi/L)
    axvline(lambda_n, color='red')
    axvline(i*pi/L, color='green')
    assert (i-1)*pi/L < lambda_n < i*pi/L

    An = -4*((L**2*h*lambda_n - (h - k)*L*lambda_n)*sin(L*lambda_n) -
h*cos(L*lambda_n) + h)/(2*L*h*lambda_n**2 + h*lambda_n*sin(2*L*lambda_n))
    T += An*cos(lambda_n*x)*exp(-alpha*lambda_n**2*t)
    flux += -1 * An * lambda_n * exp(-lambda_n**2*alpha*t)*sin(lambda_n*x)
    fluxkernel += where(t==0, 0, 
                        An*alpha*lambda_n**3*exp(-alpha*lambda_n**2*t)*sin(lambda_n*L))
legend()
xlabel('$\lambda$')
rc('font', family='serif')
rc('text', usetex=True)
#savefig('lambdacomp.pdf')
#rc('font', family='serif')
#rc('text', usetex=False)
```

```python
dt = 0.01
t = arange(0, 100, dt)
countt, = t.shape
dx = 0.01
x = np.arange(0, 0.12, dx)
countx, = x.shape

T = np.zeros([countx, countt])
T[0, 0] = 10
Q = np.ones_like(t)

#k = 237 # W/m.K
#rho = 2702 # kg/m3
#Cp = 903 # J/kg.K
#k = 386.0 # W/m.K
#rho = 8940.0 # kg/m3
#Cp = 393.5 # J/kg.K
#k = 180.0 # W/m.K
#rho = 2700.0 # kg/m3
#Cp = 880.0 # J/kg.K
k = 122.9 # W/m.K
rho = 8470.0 # kg/m3
Cp = 377.1 # J/kg.K
h = 3000 # W/m2.K
TR = 0
A = pi/4*(0.02)**2

for i in range(1,countt):
    for j in range(0,countx):
        if j==0:
            dTdxa = 0
            dTdxb = (T[j+1,i-1]-T[j,i-1])/dx
            ddTdxx = (dTdxb-dTdxa)/dx
        elif j==countx-1:
            dTdxa = (T[j,i-1]-T[j-1,i-1])/dx
            dTdxb = 0
#           dTdxb = h/k*(T[j,i-1]-TR)
            Q[i-1] = h*A*(T[j,i-1]-TR)
            ddTdxx = (dTdxb-dTdxa)/dx
        else:
            dTdxa = (T[j,i-1]-T[j-1,i-1])/dx
            dTdxb = (T[j+1,i-1]-T[j,i-1])/dx
            ddTdxx = (dTdxb-dTdxa)/dx

        T[j,i]=T[j,i-1]+((k/rho/Cp)*ddTdxx)*dt;


subplot(3, 1, 1)
plot(t,T[countx-1,:]) 
subplot(3, 1, 2)
plot(t,T[0,:]) 
subplot(3, 1, 3)
plot(t,Q) 
show()
```
