## Convective boundary condition

```python
import scipy.optimize
```

```python
alpha = 2.
L = 3
k = 4.
h = 5.

def lambda_tan(lam):
    return lam*k/h*tan(lam*L) - 1

def lambda_sin(lam):
    return lam*sin(lam*L) - h/k*cos(lam*L)

def lambda_eq(lam):
    return sin(lam*L) - h/k*cos(lam*L)/lam

mustplot = False

Nterms = 1000
t = atleast_2d(linspace(0, 10, 100)).T
x = linspace(0., L, 10)
flux = -ones((len(t), len(x)))
T = zeros((len(t), len(x))) + L - x + k/h
fluxkernel = zeros_like(t)

lambda_n = -pi/L/2.0
if mustplot:
    l = linspace(0, Nterms*pi/L, 1000)
    plot(l, lambda_tan(l), label=r'$\lambda \frac{k}{h} \tan(\lambda L) - 1$')
    plot(l, lambda_sin(l), label=r'$\lambda \sin(\lambda L) - \frac{h}{k}\cos(\lambda L)$')
    plot(l, lambda_eq(l), label=r'$\sin(\lambda L) - \frac{h}{k \lambda}\cos(\lambda L)$')
    axhline(0)
    ylim([-5, 5])

leftlim = 0
rightlim = pi/L/2.
fluxarea = 0

for i in range(1, Nterms):
    if mustplot: axvline(rightlim, color='green')
        
    lambda_n = scipy.optimize.ridder(lambda_eq, leftlim, rightlim)
    leftlim, rightlim = rightlim, rightlim + pi/L
    
    if mustplot: axvline(lambda_n, color='red')
    
    assert (i-1)*pi/L < lambda_n < i*pi/L

    An = -4/(2*L*lambda_n**2 + lambda_n*sin(2*L*lambda_n))
    T += An*cos(lambda_n*x)*exp(-alpha*lambda_n**2*t)
    flux += -1 * An * lambda_n * exp(-lambda_n**2*alpha*t)*sin(lambda_n*x)
    fluxkernel += where(t==0, 0, 
                        An*alpha*lambda_n**3*exp(-alpha*lambda_n**2*t)*sin(lambda_n*L))
    #fluxarea += where(t==0, 0,
    #              An*
    
if mustplot: 
    legend()
    xlabel('$\lambda$')
    rc('font', family='serif')
    rc('text', usetex=True)
    #savefig('lambdacomp.pdf')
    #rc('font', family='serif')
    #rc('text', usetex=False)
```
