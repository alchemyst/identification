```python
dt = 0.01
t = arange(0, 100, dt)
countt, = t.shape
dx = 0.01
x = np.arange(0, 0.12, dx)
countx, = x.shape

T = np.zeros([countx, countt])
T[0, 0] = 10
Q = np.ones_like(t)

#k = 237 # W/m.K
#rho = 2702 # kg/m3
#Cp = 903 # J/kg.K
#k = 386.0 # W/m.K
#rho = 8940.0 # kg/m3
#Cp = 393.5 # J/kg.K
#k = 180.0 # W/m.K
#rho = 2700.0 # kg/m3
#Cp = 880.0 # J/kg.K
k = 122.9 # W/m.K
rho = 8470.0 # kg/m3
Cp = 377.1 # J/kg.K
h = 3000 # W/m2.K
TR = 0
A = pi/4*(0.02)**2

for i in range(1,countt):
    for j in range(0,countx):
        if j==0:
            dTdxa = 0
            dTdxb = (T[j+1,i-1]-T[j,i-1])/dx
            ddTdxx = (dTdxb-dTdxa)/dx
        elif j==countx-1:
            dTdxa = (T[j,i-1]-T[j-1,i-1])/dx
            dTdxb = 0
#           dTdxb = h/k*(T[j,i-1]-TR)
            Q[i-1] = h*A*(T[j,i-1]-TR)
            ddTdxx = (dTdxb-dTdxa)/dx
        else:
            dTdxa = (T[j,i-1]-T[j-1,i-1])/dx
            dTdxb = (T[j+1,i-1]-T[j,i-1])/dx
            ddTdxx = (dTdxb-dTdxa)/dx

        T[j,i]=T[j,i-1]+((k/rho/Cp)*ddTdxx)*dt;


subplot(3, 1, 1)
plot(t,T[countx-1,:]) 
subplot(3, 1, 2)
plot(t,T[0,:]) 
subplot(3, 1, 3)
plot(t,Q) 
show()
```
