```python
import fit
```

The responsedata object reads response data from files and knows how to plot the response. It serves as a convenieent container for this data.

```python
data = fit.responsedata.fromfile('pulses/heat_flux.CSV')
```

```python
data.plotresponse()
```

The fft object takes a responsedata object and constructs an fft-based impulse response from it.

```python
fft = fit.fft(data, 1e-1)
```

```python
fft.bodemag()
```

We can construct a system with particular time constants using the systemwithtimeconstants object

```python
G = fit.systemwithtimeconstants([], [36])
```

```python
G
```

Compare the impulse response of the fft fit with the analytic impulse response

```python
plot(data.t, fft.impulse)
plot(*G.G.impulse())
legend(['FFT', 'Analytic'])
xlim([0, 500])
```

The fit module provides a function that will compare the data, a transfer function and an FFT

```python
fit.compare(data, G, fft)
```

Also provided is a simple fitter (using the simplex method) that will tune a response based on the sum of square error.

```python
fitter = fit.fitter(data, G)
fitter.fit()
```

```python
fitter.G
```

```python
plot(data.t, fft.impulse)
plot(*G.G.impulse())
plot(*fitter.G.G.impulse())
legend(['FFT', 'Analytic', 'Fitted'])
xlim([0, 500])
```

```python
fit.compare(data, fitter.G, fft)
```

```python
import lvm
```

```python
import glob
```

```python
rawfiles = glob.glob('raw/*.LVM')
```

```python
fitters = []
G = fit.systemwithtimeconstants([], [36])
for f in rawfiles:
    indata = lvm.lvm(f)
    data = fit.responsedata(indata.data.X_Value, indata.data.Voltage_0, indata.data.Voltage, name=indata)
    data.u -= data.u.min()
    data.y -= data.y.min()
    fitter = fit.fitter(data, G)
    fitter.fit()
    fitters.append(fitter)
```

```python
results = [(f.data.name, f.G, f.error()) for f in fitters]
results
```
